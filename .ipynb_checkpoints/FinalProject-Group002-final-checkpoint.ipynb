## Background & Prior Work

Coffee is one of the most beloved beverages in the world. The country of origin plays a significant role in determining its flavor characteristics. Optimal conditions for growing coffee include cool to warm tropical climates, rich soils, and few pests or diseases. Therefore, coffee cultivation is most favorable in tropical countries along the equator. Coffee beans have an optimal temperature range for cultivation. Prolonged exposure to temperatures outside the optimal range can deteriorate coffee bean quality and potentially lead to defects. For instance, Arabica Coffee (Coffea Arabica) is the source of roughly 70% of the world’s coffee and its optimal temperature range is 64℉-70℉[^Scott2015].


Coffee is grown in more than 50 countries. Climate, geography, processing methods, or cultural practices between countries can produce wide variations in coffee quality and commercial value. This aspect may be at risk of fraud, such as mislabeling the product to falsely attribute to a country of coffee quality. 

With the aim to maintain the authenticity of the origin of coffee beans, a study used terahertz spectroscopy combined with machine learning methods, including convolutional neural network (CNN), linear discriminant analysis (LDA), and support vector machine (SVM) to classify different geographical origins of coffee beans. The best classification results were obtained using the CNN model reaching above 90% accuracy[^Yang2021].

Although previous studies have explored the impact of coffee origin information on the sensory perception of professional coffee tasters [^Artêncio2023], models to predict the origin of a coffee bean based on its' characteristics has not been created. 

[^Scott2015]: Scott, M. (19 June 2015) Climate & Coffee. Available at: https://www.climate.gov/news-features/climate-and/climate-coffee (Accessed: July 14, 2023)

[^Yang2021]: Yang, S., et al (18 May 2021) Determination of the Geographical Origin of Coffee Beans Using Terahertz Spectroscopy Combined With Machine Learning Methods. Available at: https://www.frontiersin.org/articles/10.3389/fnut.2021.680627/full (Accessed: July 14, 2023)

[^Artêncio2023]: Artêncio, M., et al (27 March 2023) The impact of coffee origin information on sensory and hedonic judgment of fine Amazonian robusta coffee. Available at: https://onlinelibrary.wiley.com/doi/10.1111/joss.12827?af=R (Accessed: August 4, 2023)

# Hypothesis


Since the sensory attributes of coffee are affected by various environmental factors such as altitude and light exposure [^Ahmed2021], we hypothesize that it is possible to make predictions about the origin of a coffee bean based on its characteristics. 

The **null hypothesis** is that there is no relationship between the sensory attributes of coffee beans and their country of origin, which makes predictions extremely difficult.

- Our team decided that a prediction model that achieves above a **75% accuracy** would be a reasonable model. Therefore, if we establish a model with above 75% accuracy, we will reject the null hypothesis.

[^Ahmed2021]: Ahmed S., et al (08 October 2021) Climate Change and Coffee Quality: Systematic Review on the Effects of Environmental and Management Variation on Secondary Metabolites and Sensory Attributes of Coffea arabica and Coffea canephora. Available at: https://www.frontiersin.org/articles/10.3389/fpls.2021.708013/full (Accessed: July 14, 2023)

This dataset is a result of scraping the web database from the Coffee Quality Institute (CQI), a non-profit organization in the US that aims to improve the quality of coffee globally. This dataset contains sensory information (Aroma, Flavor, Aftertaste, Acidity, Body, Balance, Uniformity, Clean Cup, and Sweetness) about coffee beans from all over the world.
In addition to the sensory characteristics of the coffee beans from each country, the environmental metrics of each farm are also listed, but not used for the purposes of the model.

 With this information, our group can associate the unique features of coffee beans with the country of origin.



> The description of the sensory variables are below:

| Feature    | Description |
|---------|-----|
| Aroma    | Refers to the scent or fragrance of the coffee.  |
| Flavor    | The flavor of coffee is evaluated based on the taste, including any sweetness, bitterness, acidity, and other flavor notes.  |
| Aftertaste  | Refers to the lingering taste that remains in the mouth after swallowing the coffee.  |
| Acidity  | Acidity in coffee refers to the brightness or liveliness of the taste.  |
| Body  | The body of coffee refers to the thickness or viscosity of the coffee in the mouth.  |
| Balance  | Balance refers to how well the different flavor components of the coffee work together.  |
| Uniformity  | Uniformity refers to the consistency of the coffee from cup to cup.  |
| Clean Cup  | A clean cup refers to a coffee that is free of any off-flavors or defects, such as sourness, mustiness, or staleness.  |
| Sweetness  | It can be described as caramel-like, fruity, or floral, and is a desirable quality in coffee.  |
| Cupper Points  | Final score given by the coffee evaluator.  |
| Moisture  | Percent moisture content of the coffee bean  |

# Data Analysis & Results

**In this section, we attempt to create a model that can predict the country of origin of coffee beans based on its sensory characteristics**

> We compare the performance of various machine-learning and deep-learning models used for multi-class classification problems. 

- Logistic Regression

> Machine Learning model that is originally designed for binary classification. This model uses a sigmoid function to transform the combination of predictors into a value between 0 and 1

- Support Vector Machines (Linear Kernel, Polynomial Kernel, RBF Kernel)

> The goal of a support vector machine is to find a hyperplane that best separates the data of different groups in the feature space. In order to find a hyperplane to separate the data, support vector machines use various kernels to map the input data onto a higher-dimensional feature space.

- Neural Networks 

> Neural networks are fundamental models of deep learning. It consists of layers of multiple nodes where connections between nodes have an associated weight. Each node applies an activation function to the input and passes a transformed number to the next layer.

- Decision Tree
> A decision tree is a predicitive model that takes an input feature, and produces an output such as a label or a numerical value based on the paramaters fed into it. The classification occurs by making labelling decisions based on conditions and features of the dataset values.

- Random Forest
> A random forest classifier is an ensemble machine learning method that is used for both classification and regression analyses. It makes use of multiple decision trees and random sampling for training in order to build a more robust classification system.

- K Nearest Neighbors
> The K-Nearest Neighbors algorithm is a predicitve algorithm that bases a prediction on similar points between training data and input data.

```python
# Here, we set up some arrays to store the average accuracy of every model we try
model_types = []
accuracy_models = []
```

### Logistic Regression

```python
# Checking model performance with 5-fold cross validation 

# separate features and target
X = coffee_filtered.drop(['country', 'cupper_points'], axis=1)  # Features
Y = coffee_filtered["country"]  # Target 

# define model
model = LogisticRegression() 

# define number of cv folds
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# cross-validation and get the accuracy scores for each fold
accuracy_scores = cross_val_score(model, X, Y, cv=kfold, scoring='accuracy')

#take the average of 5 fold
average_accuracy = np.mean(accuracy_scores)

print(f"Average Accuracy Logistic: {average_accuracy}")

#populate results 
model_types.append("Logistic")
accuracy_models.append(average_accuracy)
```

### Support Vector Machines

```python
#decide which features we want to use to predict the country of origin 
print(coffee_filtered.columns)

#separate features and target
X = coffee_filtered.drop(['country', 'cupper_points'], axis=1)  
Y = coffee_filtered["country"]
```

**Linear Kernel SVM**

```python
# linear kernel SVM
support_vector_linear = SVC(kernel='linear')

#define k-fold
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# cross-validation and get the accuracy scores for each fold
accuracy_scores = cross_val_score(support_vector_linear, X, Y, cv=kfold, scoring='accuracy')

#average accuracy
average_accuracy = np.mean(accuracy_scores)

print(f"Average Accuracy SVC Linear: {average_accuracy}")

#populate results
model_types.append("SVC_Linear")
accuracy_models.append(average_accuracy)
```

**Polynomial Kernel SVM**

```python
# polynomial kernel SVM
support_vector_poly = SVC(kernel='poly', degree=3, gamma='auto')

#define k-fold
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# cross-validation and get the accuracy scores for each fold
accuracy_scores = cross_val_score(support_vector_poly, X, Y, cv=kfold, scoring='accuracy')

#average accuracy
average_accuracy = np.mean(accuracy_scores)

print(f"Average Accuracy SVC Poly: {average_accuracy}")

#populate results
model_types.append("SVC_Poly")
accuracy_models.append(average_accuracy)
```

> We note that we could fine-tune the "degree" parameter of this polynomial kernel svm to achieve a higher accuracy. Trying multiple values for the "degree" parameter is computationally expensive, so we decided to stick with the value 3.

**RBF Kernel SVM**

```python
regularization = [i for i in range(1, 26)]
accuracy_tuning = []

#tune regularization parameter
for r in regularization:

    # polynomial kernel SVM
    support_vector_rbf = SVC(kernel='rbf', gamma='auto', C=r, random_state=42)

    #define k-fold
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)

    # cross-validation and get the accuracy scores for each fold
    accuracy_scores = cross_val_score(support_vector_rbf, X, Y, cv=kfold, scoring='accuracy')

    #average accuracy
    average_accuracy = np.mean(accuracy_scores)
    
    accuracy_tuning.append(average_accuracy)

best_accuracy = max(accuracy_tuning)
best_r = accuracy_tuning.index(best_accuracy)

print(f"Average Accuracy SVC RBF: {best_accuracy}, with regularization parameter: {best_r}")

#populate results
model_types.append("SVC_RBF")
accuracy_models.append(best_accuracy)
```

> Here, we fine-tuned the regularization parameter to optimize our RBF SVM model.

### Neural Networks

```python
# define a function to build a neural network 
# adjust this function to add layers, or increase the number of nodes in each layer

def craft_model(i_shape, n_categories):
    model = keras.Sequential([
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(32, activation='relu'),
        keras.layers.Dense(n_categories, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
#separate features and target
X = coffee_filtered.drop(['country', 'cupper_points'], axis=1)  
Y = coffee_filtered["country"]  
```

```python
#encode the categories into numerical values
encoder = LabelEncoder()
Y_encoded = encoder.fit_transform(Y)
num_categories = len(encoder.classes_)

#change Y into one-hot encoding
Y_categorical = to_categorical(Y_encoded, num_categories)

#define input shape
i_shape = X.shape[1:]

#define cross validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

#store accuracy here during cv
accuracy_list = []
```

```python
# cv loop
for train_idex, test_idex in kfold.split(X):
    X_train, X_test = X.iloc[train_idex], X.iloc[test_idex]
    y_train, y_test = Y_categorical[train_idex], Y_categorical[test_idex]

    # use function to build model
    model = craft_model(i_shape, num_categories)
    
    early_stop = EarlyStopping(monitor='accuracy', patience=5, restore_best_weights=True)

    # train the model with training split
    model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0, callbacks=[early_stop])

    # Evaluate the model on the test set
    _, accuracy = model.evaluate(X_test, y_test, verbose=0)
    accuracy_list.append(accuracy)

average_accuracy = np.mean(accuracy_list)
    
print(f"Average Accuracy Neural Network: {average_accuracy}")

#populate results
model_types.append("Neural_Network")
accuracy_models.append(average_accuracy)
```

### Random Forest

```python
# Create a Random Forest classifier
rf = RandomForestClassifier()

# Define a range of hyperparameters to search through
param_grid = {
    'n_estimators': [100, 200, 300],          # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],          # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],          # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]             # Minimum number of samples required to be at a leaf node
}

# Create the GridSearchCV object
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='accuracy')

# Fit the model with different hyperparameters
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Train the classifier with the best hyperparameters
rf = RandomForestClassifier(**best_params)
rf.fit(X_train, y_train)

#define k-fold
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# cross-validation and get the accuracy scores for each fold
accuracy_scores = cross_val_score(rf, X, Y, cv=kfold, scoring='accuracy')

#average accuracy
average_accuracy = np.mean(accuracy_scores)

print(f"Average Accuracy Random Forest: {average_accuracy}")

#populate results
model_types.append("Random_Forest")
accuracy_models.append(average_accuracy)
```

### Decision Tree

```python
# Create a Decision Tree classifier
dt = DecisionTreeClassifier()

# Define a range of hyperparameters to search through
param_grid = {
    'criterion': ['gini', 'entropy'],   # The function to measure the quality of a split
    'max_depth': [None, 10, 20, 30],    # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],    # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]       # Minimum number of samples required to be at a leaf node
}

# Create the GridSearchCV object
grid_search = GridSearchCV(dt, param_grid, cv=5, scoring='accuracy')

# Fit the model with different hyperparameters
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Train the classifier with the best hyperparameters
dt = DecisionTreeClassifier(**best_params)
dt.fit(X_train, y_train)

#define k-fold
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# cross-validation and get the accuracy scores for each fold
accuracy_scores = cross_val_score(dt, X, Y, cv=kfold, scoring='accuracy')

#average accuracy
average_accuracy = np.mean(accuracy_scores)

print(f"Average Accuracy Decision Tree: {average_accuracy}")

#populate results
model_types.append("Decision Tree")
accuracy_models.append(average_accuracy)
```

### K Nearest Neighbors

```python
# Create a KNN classifier
knn = KNeighborsClassifier()

# Define a range of hyperparameters to search through
param_grid = {
    'n_neighbors': range(1, 11),        # Number of neighbors to consider
    'weights': ['uniform', 'distance'], # Weight function used in prediction
    'p': [1, 2]                         # Power parameter for the Minkowski metric (1 for Manhattan distance, 2 for Euclidean distance)
}

# Create the GridSearchCV object
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')

# Fit the model with different hyperparameters
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Train the classifier with the best hyperparameters
knn = KNeighborsClassifier(**best_params)
knn.fit(X_train, y_train)

#define k-fold
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# cross-validation and get the accuracy scores for each fold
accuracy_scores = cross_val_score(knn, X, Y, cv=kfold, scoring='accuracy')

#average accuracy
average_accuracy = np.mean(accuracy_scores)

#populate results
model_types.append("KNN")
accuracy_models.append(average_accuracy)
```

**so many different measurements for the altitude! Let's try to make most of these a single number in meters**

```python
# function to clean the altitude measurements 
def feet_meters(feet_str):
    return float(feet_str) * 0.3048

def clean_altitude(altitude_str):
    patterns = [
        r'(\d+)-(\d+)',                 # Range: 1950-2200
        r'~(\d+)',                      # Approximate: ~350m
        r'(\d+)\s*-\s*(\d+)\s*m',       # Range with 'm': 1600 - 1800 m
        r'(\d+)\s*meters?',             # Meters: 1100 meters
        r'(\d+\.\d+)\s*m',              # Decimal meters: meters above sea level: 1.872
        r'(\d+)\s*ft',                  # Feet to meters: 2000 ft
        r'(\d+)-(\d+)\s*ft',            # Feet range to meters: 3600-6200 ft 
        r'(\d+)\s*msnm',                # Meters above sea level (msnm): 1300 msnm
        r'(\d+)~(\d+)\s*m\.?s\.?l',     # Tilde separated meters above sea level: 1300~1400 m.s.l
        r'(\d+)-(\d+)\s*metros',        # Range in meters: 5600-5760 metros
        r'above\s*(\d+)\s*m',           # Above meters: above 1600 m
        r'(\d+)\s*masl',                # Meters above sea level (masl): 1400 masl
        r'de\s*(\d+)\s*a\s*(\d+)\s*msnm',   # Range in meters above sea level: de 1600 a 1950 msnm
        r'(\d+)\+\+',                   # '++' after value: 800++
        r'(\d+)-(\d+)\s*psnm',          # Range in pseudo-meters above sea level: 5500-6000 psnm
        r'(\d+)\s*m\.?s\.?n\.?m\.?',    # Meters above sea level with periods: 1400 m.s.n.m.
        r'(\d+)-(\d+)\s*masl',          # Range in meters above sea level with masl: 1200-1600masl
        r'(\d+)\s*psn',                 # Pseudo-meters above sea level: 4000psn
        r'de\s*(\d+)\s*a\s*(\d+)\s*msn', # Range in meters above sea level with 'msn': de 1.600 a 1.950 msn
        r'(\d+)-(\d+)\s*metros',        # Range in meters: 5440-5760 metros
        r'(\d+)\s*m公尺',                # Meters in a non-English script (e.g., Chinese): 700m公尺
        r'(\d+)\s*a\s*(\d+)\s*mtrs',     # Range in meters with 'mtrs': 1300 a 1800 mtrs
        r'approx\.\s*(\d+)[\’\']*(\d*)\s*metres?\s*a\.?s\.?l\.?',  # Approximate meters above sea level: approx. 1'500 metres a.s.l.
    ]

    # find matching patterns
    for pattern in patterns:
        matches = re.findall(pattern, altitude_str)
        if matches:
            # if there is a range, average it
            if len(matches[0]) == 2:
                lower_val, upper_val = map(float, matches[0])
                # convert feet to meters
                if "ft" in altitude_str.lower() or "feet" in altitude_str.lower():
                    lower_val = feet_meters(lower_val)
                    upper_val = feet_meters(upper_val)
                return (lower_val + upper_val) / 2
            else:
                value = float(matches[0][0])
                # convert ft or feet to meters
                if "ft" in altitude_str.lower() or "feet" in altitude_str.lower():
                    value = feet_meters(value)
                return value

    # no pattern? just return none
    return None
```
